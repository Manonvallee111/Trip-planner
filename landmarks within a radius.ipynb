```python
now = datetime.now()
directions_result = gmaps.directions(origin, destination, mode="driving", departure_time=now)
```

```python
route = polyline.decode(directions_result[0]['overview_polyline']['points'])
```

```python
landmarks = []
for point in route:
    # Perform a nearby search for landmarks within a 70km radius
    places_result = gmaps.places_nearby(location=point, radius=20000, type="tourist_attraction")

    # Loop through the results and add them to the list of landmarks
    for place in places_result['results']:
        landmarks.append(place['name'])

# Print the list of landmarks
print(landmarks)
```
